```python
# cleaning rawg

rawg_clean = rawg_data.drop(['website', 'tba', 'metacritic'], axis =1)
rawg_clean = rawg_clean.dropna()
rawg_clean
```

```python
print("Sorted by Number Owned")
# This is for observing the data. Changing this won't change what is written in filter_rawg_info.csv

rawg_filtered = rawg_clean[["name", "rating", "playtime", "ratings_count", "added_status_owned"]]
# To add any new columns you want to observe, add the column name into the string list above

rawg_owned = rawg_filtered.sort_values(by=["added_status_owned"], ascending=False)

rawg_owned
```

```python
print("Sorted by Rating Count")
rawg_rc = rawg_filtered.sort_values(by=["ratings_count"], ascending=False)

rawg_rc
```

```python
print("Sorted by Playtime")
rawg_playtime = rawg_filtered.sort_values(by=["playtime"], ascending=False)

rawg_playtime
```

```python
print("Sorted by Rating")
rawg_rating = rawg_filtered.sort_values(by=["rating"], ascending=False)

rawg_rating
```

```python
# Writing filtered DF to filtered_rawg_info.csv. This one is NOT sorted by any specific column.

# To alter written data in csv file, add/remove/change strings to the list of column names you're interested in
rawg_filtered_csv = rawg_clean[["id", "name", "rating", "playtime", "ratings_count", "added_status_owned"]]
# Current Columns: id, name, rating, playtime, ratings count, how many owned

rawg_filtered_csv.to_csv("filtered_rawg_info.csv", index=False) # index=False removes row numbers, index=True keeps row numbers
```
